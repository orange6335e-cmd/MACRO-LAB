```python
# 1. 掛載 Google 雲端硬碟
from google.colab import drive
drive.mount('/content/drive')

import time
import torch
from torch.utils.data import Dataset, DataLoader
import pandas as pd
import numpy as np

# 2. 讀取特徵檔案 (請確認路徑跟你存在雲端硬碟的資料夾一致)
file_path = '/content/drive/MyDrive/DeltaLag_Project/deltalag_features_top300.parquet'
print("正在讀取 Parquet 檔案...")
df = pd.read_parquet(file_path)
print(f"資料讀取成功！總筆數: {df.shape[0]:,}")

# 3. 定義高速版 DeltaLag Dataset（用 3D NumPy 矩陣取代逐檔 Pandas 查詢）
class DeltaLagDataset(Dataset):
    def __init__(self, df, L=20, l_max=10):
        self.L = L
        self.l_max = l_max

        # 建立日期與股票的索引字典
        self.dates = sorted(df['Date'].unique())
        self.tickers = sorted(df['Ticker'].unique())
        self.date_to_idx = {d: i for i, d in enumerate(self.dates)}
        self.ticker_to_idx = {t: i for i, t in enumerate(self.tickers)}

        num_dates = len(self.dates)
        num_stocks = len(self.tickers)
        self.feature_cols = [
            'Close', 'Volume', 'Open_Close_Ratio',
            'High_Close_Ratio', 'Low_Close_Ratio',
            'Log_Volume', 'Daily_Return'
        ]

        # 建立 3D 矩陣 (日期, 股票, 特徵)，預設 NaN
        print("🚀 正在將 Pandas 轉換為極速 3D 矩陣...")
        start_time = time.time()

        self.X_all = np.full((num_dates, num_stocks, len(self.feature_cols)), np.nan, dtype=np.float32)
        self.y_all = np.full((num_dates, num_stocks), np.nan, dtype=np.float32)

        # 向量化填入資料（確保索引為整數，避免 datetime 型別不一致產生 NaN）
        date_indices = df['Date'].map(self.date_to_idx).astype(int).values
        ticker_indices = df['Ticker'].map(self.ticker_to_idx).astype(int).values

        self.X_all[date_indices, ticker_indices, :] = df[self.feature_cols].values
        self.y_all[date_indices, ticker_indices] = df['Target_Return'].values

        print(f"✅ 轉換完成！耗時: {time.time() - start_time:.2f} 秒")

        self.valid_dates_idx = list(range(L - 1, num_dates))

    def __len__(self):
        return len(self.valid_dates_idx)

    def __getitem__(self, idx):
        target_date_idx = self.valid_dates_idx[idx]

        # 一次切出過去 L 天全市場資料 -> (L, Stocks, Features)
        X_window = self.X_all[target_date_idx - self.L + 1 : target_date_idx + 1, :, :]
        y_target = self.y_all[target_date_idx, :]

        # 轉置為 (Stocks, L, Features)
        X_window = np.transpose(X_window, (1, 0, 2))

        # 過濾掉 20 天內有任何 NaN 的股票
        valid_mask = ~np.isnan(X_window).any(axis=(1, 2)) & ~np.isnan(y_target)
        X_valid = X_window[valid_mask]
        y_valid = y_target[valid_mask]

        return torch.tensor(X_valid, dtype=torch.float32), torch.tensor(y_valid, dtype=torch.float32)

# 4. 建構 DataLoader
print("\n正在切割 Train / Val / Test 資料集...")

# 1. 訓練集 (Train): 2014 ~ 2020 年底
train_df = df[(df['Date'] >= '2014-01-01') & (df['Date'] <= '2020-12-31')]

# 2. 驗證集 (Validation): 2021 ~ 2022 年底 (用來調校超參數)
val_df = df[(df['Date'] >= '2021-01-01') & (df['Date'] <= '2022-12-31')]

# 3. 測試集 (Test): 2023 ~ 2026 年 (終極盲測，放進保險箱最後才能看)
test_df = df[df['Date'] >= '2023-01-01']

# 4. 完整訓練集 (Full Train): 2014 ~ 2022 年底 (找到最佳參數後，用這份重新訓練)
full_train_df = df[(df['Date'] >= '2014-01-01') & (df['Date'] <= '2022-12-31')]

print(f"📊 訓練集 (Train 2014-2020) 筆數: {len(train_df):,}")
print(f"📊 驗證集 (Val 2021-2022) 筆數: {len(val_df):,}")
print(f"📊 完整訓練集 (Full Train 2014-2022) 筆數: {len(full_train_df):,}")
print(f"📊 測試集 (Test 2023~) 筆數: {len(test_df):,}")

# 建立對應的 Dataset 與 DataLoader
print("\n🚀 正在建構極速 3D 張量 (這會花幾秒鐘)...")
train_dataset = DeltaLagDataset(train_df, L=20, l_max=10)
val_dataset = DeltaLagDataset(val_df, L=20, l_max=10)
full_train_dataset = DeltaLagDataset(full_train_df, L=20, l_max=10)
test_dataset = DeltaLagDataset(test_df, L=20, l_max=10)

# batch_size=1 代表一次拿「一天全市場」的橫截面資料
train_loader = DataLoader(train_dataset, batch_size=1, shuffle=False)
val_loader = DataLoader(val_dataset, batch_size=1, shuffle=False)
full_train_loader = DataLoader(full_train_dataset, batch_size=1, shuffle=False)
test_loader = DataLoader(test_dataset, batch_size=1, shuffle=False)

print("✅ 所有 DataLoader 準備完畢！")

# 抽一個 Batch 來看形狀，並推上 GPU
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
x_batch, y_batch = next(iter(train_loader))
x_batch, y_batch = x_batch.to(device), y_batch.to(device)
print(f"\n✅ 測試成功！使用硬體: {device}")
print(f"X_batch shape: {x_batch.shape} --> (Batch, 股票數量, 過去{train_dataset.L}天, 特徵數量)")
print(f"y_batch shape: {y_batch.shape} --> (Batch, 股票數量)")
```

```python
import torch
import torch.nn as nn
import torch.nn.functional as F

class DeltaLagNet(nn.Module):
    def __init__(self, num_features=7, hidden_dim=32, l_max=10, k=2):
        super(DeltaLagNet, self).__init__()
        self.l_max = l_max
        self.k = k
        self.hidden_dim = hidden_dim

        # 1. 時間序列編碼器 (Temporal Encoder)
        # 用來理解每一檔股票過去 20 天的走勢
        self.lstm = nn.LSTM(input_size=num_features, hidden_size=hidden_dim, batch_first=True)

        # 2. 注意力機制的投影矩陣 (W_Q, W_K)
        self.W_Q = nn.Linear(hidden_dim, hidden_dim, bias=False)
        self.W_K = nn.Linear(hidden_dim, hidden_dim, bias=False)

        # 3. 預測層 (MLP)
        # 輸入維度是 num_features (因為我們抓的是原始特徵)，輸出 1 個預測報酬率
        self.mlp = nn.Sequential(
            nn.Linear(num_features, 16),
            nn.ReLU(),
            nn.Linear(16, 1)
        )

    def forward(self, x):
        # x 原始形狀: (Batch=1, N_stocks=230, L=20, Features=7)
        # 因為我們是橫截面一天一天算，直接把 Batch 維度壓縮掉
        x = x.squeeze(0) # 變成 (N, L, F)
        N, L, F_dim = x.shape

        # --- 階段一：LSTM 特徵抽取 ---
        # lstm_out 形狀: (N, L, hidden_dim)
        lstm_out, _ = self.lstm(x)

        # --- 階段二：建構 Query 與 Key ---
        # Query (目標股票): 只取最後一天的隱藏狀態 h_L
        h_L = lstm_out[:, -1, :] # (N, hidden_dim)
        Q = self.W_Q(h_L)        # (N, hidden_dim)

        # Key (潛在領先股): 取過去 l_max 天的隱藏狀態
        h_lmax = lstm_out[:, L - self.l_max : L, :] # (N, l_max, hidden_dim)
        K = self.W_K(h_lmax)                        # (N, l_max, hidden_dim)

        # --- 階段三：計算 Cross-Attention 分數 ---
        # 利用愛因斯坦求和約定 (einsum) 一次把 230x230 檔股票的矩陣乘法算完
        # A 形狀: (N_目標, N_領先, l_max)
        A = torch.einsum('ud, vjd -> uvj', Q, K)

        # 論文重點：屏除自我領先 (Self-Attention)
        # 我們創造一個對角線遮罩，把目標股票等於領先股票的分數設為極小值(-inf)
        mask = torch.eye(N, device=x.device).unsqueeze(-1).expand(N, N, self.l_max).bool()
        A = A.masked_fill(mask, float('-inf'))

        # --- 階段四：Top-K 動態尋找 Leader 與 Lag 天數 ---
        # 將注意力矩陣攤平來找前 K 高分
        A_flat = A.view(N, -1) # 形狀: (N, N * l_max)
        topk_scores, topk_indices = torch.topk(A_flat, self.k, dim=1) # (N, k)

        # 將分數通過 Softmax 轉成權重
        attn_weights = F.softmax(topk_scores, dim=1) # (N, k)

        # 解碼攤平的 index，找出是哪檔股票 (peer_idx) 跟延遲了幾天 (lag_idx)
        peer_indices = torch.div(topk_indices, self.l_max, rounding_mode='floor') # 股票索引 (N, k)
        lag_indices = topk_indices % self.l_max                                   # 延遲索引 (N, k)

        # --- 階段五：提取原始特徵並聚合 ---
        # 還原到真實的時間軸 index (0~19)
        temporal_indices = L - self.l_max + lag_indices

        # 根據算出來的 peer 跟 time，直接去原本的輸入 x 抓取「原始特徵」
        z_raw = x[peer_indices, temporal_indices, :] # 形狀: (N, k, F_dim)

        # 將 K 檔領先股的特徵，依照 Attention 權重加總
        # attn_weights.unsqueeze(-1) 是為了讓形狀對齊 (N, k, 1) 以進行廣播相乘
        z_aggregated = torch.sum(z_raw * attn_weights.unsqueeze(-1), dim=1) # 形狀: (N, F_dim)

        # --- 階段六：MLP 預測 ---
        preds = self.mlp(z_aggregated) # (N, 1)

        # 把維度壓平回傳 (N,)
        return preds.squeeze(-1)

# --- 鎖定隨機種子（必須在模型初始化之前，才能固定權重） ---
import random
import numpy as np
SEED = 42
random.seed(SEED)
np.random.seed(SEED)
torch.manual_seed(SEED)
torch.cuda.manual_seed_all(SEED)
torch.backends.cudnn.deterministic = True
torch.backends.cudnn.benchmark = False

# --- 試運轉 (Test Run) ---
print("初始化 DeltaLag 模型...")
# 將模型推上 GPU
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = DeltaLagNet(num_features=7, hidden_dim=32, l_max=10, k=2).to(device)

# 拿我們剛剛的 x_batch 餵進去試試看
model.eval() # 測試模式
with torch.no_grad():
    predictions = model(x_batch)

print(f"模型輸出形狀: {predictions.shape} --> 預期應該要是 (230,)")
print(f"前五檔股票的預測次日報酬率 (未經訓練的亂數): \n{predictions[:5].cpu().numpy()}")
```

```python
import torch.nn.functional as F

class MonotonicRankLoss(nn.Module):
    def __init__(self):
        super(MonotonicRankLoss, self).__init__()

    def forward(self, preds, targets):
        """
        preds: 模型預測出來的報酬率，形狀 (N,)
        targets: 隔日的實際報酬率，形狀 (N,)
        """
        # 將形狀轉為 (N, 1) 以便進行矩陣廣播
        preds = preds.view(-1, 1)
        targets = targets.view(-1, 1)

        # 1. 計算所有股票對 (Pairwise) 的差異矩陣，形狀會自動擴張成 (N, N)
        # diff_targets[i, j] 就代表: 真實世界中，股票 i 減去 股票 j 的報酬差
        diff_targets = targets - targets.T
        diff_preds = preds - preds.T

        # 2. 套用 tanh 函數：這能把極端的報酬差異平滑化到 -1 到 1 之間，
        # 只保留「誰大誰小」的單調性方向
        interaction = -torch.tanh(diff_targets) * torch.tanh(diff_preds)

        # 3. 論文公式的 log(1 + exp(x))
        # 在 PyTorch 裡，我們直接呼叫 F.softplus(x)，這不僅數學上等價，而且能防止數值溢位 (NaN)
        loss_matrix = F.softplus(interaction)

        # 4. 取平均回傳
        # (對角線自己比自己的 loss 是常數 softplus(0)=0.693，取平均不會影響梯度下降方向)
        return loss_matrix.mean()

# --- 試運轉 Loss ---
print("測試 MonotonicRankLoss...")
loss_fn = MonotonicRankLoss()

# 拿剛剛瞎猜的 predictions 跟真實的 y_batch 算算看誤差
test_loss = loss_fn(predictions, y_batch.squeeze(0))
print(f"初始未訓練的 Loss 值: {test_loss.item():.4f}")
```

```python
import torch
import numpy as np
import pandas as pd
from scipy.stats import spearmanr
from tqdm.auto import tqdm

def evaluate_model_performance(model, dataloader, is_train=True):
    model.eval() # 切換到評估模式，關閉 Dropout 與 BatchNorm

    daily_ic = []
    daily_strategy_returns = []
    daily_benchmark_returns = []

    print(f"📊 開始計算{'訓練集 (In-Sample)' if is_train else '測試集 (Out-of-Sample)'} 績效指標...")

    device = next(model.parameters()).device

    with torch.no_grad(): # 評估階段不需要計算梯度，可以省記憶體並加速
        pbar = tqdm(dataloader, desc="Evaluating")
        for x_batch, y_batch in pbar:
            x_batch, y_batch = x_batch.to(device), y_batch.to(device)
            y_target = y_batch.squeeze(0).cpu().numpy()

            # --- Z-score 標準化 (必須跟訓練時一模一樣，在 dim=(1,2) 即股票與時間維度) ---
            x_mean = x_batch.mean(dim=(1, 2), keepdim=True)
            x_std = x_batch.std(dim=(1, 2), keepdim=True) + 1e-8
            x_batch_norm = (x_batch - x_mean) / x_std

            # --- 進行預測 ---
            preds = model(x_batch_norm).cpu().numpy()

            # --- 1. 計算 IC (Information Coefficient) ---
            # 計算橫截面上，預測排名與實際排名的斯皮爾曼等級相關係數
            ic, _ = spearmanr(preds, y_target)
            if not np.isnan(ic):
                daily_ic.append(ic)

            # --- 2. 模擬純做多投資組合 (Top 10% Long Only) ---
            num_stocks = len(preds)
            k = max(1, int(num_stocks * 0.1)) # 取前 10% 的數量

            # 找出預測分數最高的股票 index (做多)
            sorted_indices = np.argsort(preds)
            long_indices = sorted_indices[-k:]     # 分數最高的 k 檔

            # 計算當日策略報酬：做多部位的平均報酬
            long_return = np.mean(y_target[long_indices])

            # 計算當日市場基準報酬（等權重全市場平均）
            market_return = np.mean(y_target)

            if not np.isnan(long_return):
                daily_strategy_returns.append(long_return)
                daily_benchmark_returns.append(market_return)

    # --- 3. 統整並計算年化指標 (假設一年 252 個交易日) ---
    mean_ic = np.mean(daily_ic)

    daily_returns_array = np.array(daily_strategy_returns)
    daily_benchmark_array = np.array(daily_benchmark_returns)

    mean_daily_return = np.mean(daily_returns_array)
    std_daily_return = np.std(daily_returns_array)

    mean_daily_benchmark = np.mean(daily_benchmark_array)

    # 超額報酬 = 策略報酬 - 市場基準報酬
    daily_excess = daily_returns_array - daily_benchmark_array
    mean_daily_excess = np.mean(daily_excess)
    std_daily_excess = np.std(daily_excess)

    # 年化報酬率 (AR) = 日均報酬 * 252
    annualized_return = mean_daily_return * 252
    annualized_benchmark = mean_daily_benchmark * 252
    annualized_excess = mean_daily_excess * 252

    # 夏普值 (SR) = 日均報酬 / 日報酬標準差 * sqrt(252)
    sharpe_ratio = (mean_daily_return / std_daily_return) * np.sqrt(252) if std_daily_return > 0 else 0

    # 資訊比率 (IR) = 超額報酬 / 超額報酬標準差 * sqrt(252)
    info_ratio = (mean_daily_excess / std_daily_excess) * np.sqrt(252) if std_daily_excess > 0 else 0

    print("\n" + "="*50)
    print(f"📈 {'訓練集' if is_train else '測試集'} 綜合表現 (純做多策略 Top 10%):")
    print(f"平均 IC (Information Coefficient): {mean_ic:.4f}")
    print(f"策略年化報酬率: {annualized_return * 100:.2f}%")
    print(f"市場基準年化報酬率: {annualized_benchmark * 100:.2f}%")
    print(f"超額年化報酬率 (Alpha): {annualized_excess * 100:.2f}%")
    print(f"夏普值 (Sharpe Ratio): {sharpe_ratio:.2f}")
    print(f"資訊比率 (Information Ratio): {info_ratio:.2f}")
    print("="*50 + "\n")

    return mean_ic, annualized_return, sharpe_ratio

# --- 執行評估 ---
# 我們拿剛剛高速版的 train_loader 餵給模型
evaluate_model_performance(model, train_loader, is_train=True)
```

```python
!pip install optuna
import optuna
import torch.optim as optim

def objective(trial):
    # 1. 讓 Optuna 幫我們「擲骰子」決定這回合的參數
    hidden_dim = trial.suggest_categorical("hidden_dim", [16, 32, 64, 128])
    l_max = trial.suggest_categorical("l_max", [3, 5, 10, 15])
    k = trial.suggest_categorical("k", [1, 2, 3, 5, 8])
    lr = trial.suggest_float("lr", 1e-4, 5e-3, log=True)
    weight_decay = trial.suggest_float("weight_decay", 1e-5, 1e-3, log=True)
    epochs = trial.suggest_categorical("epochs", [3, 5, 8])

    print(f"\n🎲 嘗試參數組合: hidden_dim={hidden_dim}, l_max={l_max}, k={k}, "
          f"lr={lr:.5f}, wd={weight_decay:.5f}, epochs={epochs}")

    # 2. 建立新模型與優化器
    model = DeltaLagNet(num_features=7, hidden_dim=hidden_dim, l_max=l_max, k=k).to(device)
    optimizer = optim.AdamW(model.parameters(), lr=lr, weight_decay=weight_decay)
    loss_fn = MonotonicRankLoss()

    # 3. 訓練模型
    for epoch in range(epochs):
        model.train()
        for x_batch, y_batch in train_loader:
            x_batch, y_batch = x_batch.to(device), y_batch.to(device)

            x_mean = x_batch.mean(dim=(1, 2), keepdim=True)
            x_std = x_batch.std(dim=(1, 2), keepdim=True) + 1e-8
            x_batch_norm = (x_batch - x_mean) / x_std

            optimizer.zero_grad()
            preds = model(x_batch_norm)
            loss = loss_fn(preds, y_batch.squeeze(0))
            loss.backward()
            torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm=1.0)
            optimizer.step()

        # Optuna 剪枝：每個 epoch 結束後檢查表現，太差就提早終止
        if epoch >= 1:  # 至少跑完 2 個 epoch 再判斷
            _, _, interim_sharpe = evaluate_model_performance(model, val_loader, is_train=False)
            trial.report(interim_sharpe, epoch)
            if trial.should_prune():
                print(f"  ✂️ 第 {epoch+1} 輪被剪枝，這組參數不夠好，提早終止。")
                raise optuna.exceptions.TrialPruned()

    # 4. 在「驗證集 (Val)」上盲測這組參數的實力
    _, _, val_sharpe = evaluate_model_performance(model, val_loader, is_train=False)

    print(f"  📊 本次驗證集夏普值: {val_sharpe:.2f}")
    return val_sharpe

# --- 啟動尋找黃金參數的旅程 ---
print("🚀 啟動 Optuna 超參數最佳化引擎...")
print("   策略：純做多 Top 10%，目標最大化驗證集夏普值\n")

# direction="maximize" 代表我們希望夏普值越大越好
# MedianPruner 會自動剪掉表現低於中位數的 trial，節省大量時間
study = optuna.create_study(
    direction="maximize",
    pruner=optuna.pruners.MedianPruner(n_startup_trials=3, n_warmup_steps=1)
)

# n_trials=20 代表我們給 AI 20 次機會去嘗試不同組合
study.optimize(objective, n_trials=20)

# --- 輸出最佳結果 ---
print("\n" + "="*50)
print("🏆 最佳參數組合出爐：")
print("="*50)
for key, value in study.best_params.items():
    print(f"  {key}: {value}")
print(f"\n🥇 驗證集最高夏普值: {study.best_value:.2f}")
print("="*50)

# --- 用最佳參數重新訓練完整模型 (2014~2022) ---
print("\n🔄 使用最佳參數，在 2014~2022 完整資料上重新訓練模型...")
print("   (將訓練集 + 驗證集合併，最大化模型學到的資訊量)\n")
best = study.best_params

model = DeltaLagNet(
    num_features=7,
    hidden_dim=best["hidden_dim"],
    l_max=best["l_max"],
    k=best["k"]
).to(device)

optimizer = optim.AdamW(model.parameters(), lr=best["lr"], weight_decay=best["weight_decay"])
loss_fn = MonotonicRankLoss()

for epoch in range(best["epochs"]):
    model.train()
    total_loss = 0.0
    for x_batch, y_batch in full_train_loader:  # 使用 2014~2022 完整訓練集
        x_batch, y_batch = x_batch.to(device), y_batch.to(device)
        x_mean = x_batch.mean(dim=(1, 2), keepdim=True)
        x_std = x_batch.std(dim=(1, 2), keepdim=True) + 1e-8
        x_batch_norm = (x_batch - x_mean) / x_std

        optimizer.zero_grad()
        preds = model(x_batch_norm)
        loss = loss_fn(preds, y_batch.squeeze(0))
        loss.backward()
        torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm=1.0)
        optimizer.step()
        total_loss += loss.item()

    avg_loss = total_loss / len(full_train_loader)
    print(f"  Epoch {epoch+1}/{best['epochs']} | Loss: {avg_loss:.4f}")

print("\n✅ 最佳模型訓練完成！開始終極盲測...")

# 測試集績效 — 2023 以後的資料，模型從未見過
print("\n📊 測試集 (2023~2026) 績效 (終極盲測)：")
evaluate_model_performance(model, test_loader, is_train=False)
```

```python
# === 加入手續費的績效評估（當沖策略成本） ===
# 手續費 2.8 折：買進 0.03990% + 賣出 0.03990% + 當沖證交稅 0.15% = 單趟 0.2298%
# 因為我們每天換股（隔日報酬），所以每天都會產生一次買賣成本

import torch
import numpy as np
from scipy.stats import spearmanr
from tqdm.auto import tqdm

# --- 手續費參數設定 ---
COMMISSION_RATE = 0.001425   # 券商手續費率 0.1425%
COMMISSION_DISCOUNT = 0.28   # 手續費折扣 2.8 折
DAY_TRADE_TAX = 0.0015       # 當沖證交稅 0.15%

# 每日總交易成本（買進手續費 + 賣出手續費 + 當沖證交稅）
DAILY_COST = COMMISSION_RATE * COMMISSION_DISCOUNT * 2 + DAY_TRADE_TAX  # ≈ 0.002298

print(f"📋 交易成本設定：")
print(f"   券商手續費: {COMMISSION_RATE*100:.4f}% × {COMMISSION_DISCOUNT*100:.0f}% 折 = {COMMISSION_RATE*COMMISSION_DISCOUNT*100:.4f}% (單邊)")
print(f"   當沖證交稅: {DAY_TRADE_TAX*100:.2f}%")
print(f"   每日總成本: {DAILY_COST*100:.4f}%\n")


def evaluate_with_fee(model, dataloader, daily_cost=DAILY_COST):
    """
    加入手續費後的純做多策略績效評估
    每天選 Top 10% 股票做多，隔天全部賣出，扣除當沖交易成本
    """
    model.eval()

    daily_ic = []
    daily_gross_returns = []      # 未扣手續費
    daily_net_returns = []        # 已扣手續費
    daily_benchmark_returns = []  # 市場基準

    device = next(model.parameters()).device

    with torch.no_grad():
        pbar = tqdm(dataloader, desc="Evaluating (含手續費)")
        for x_batch, y_batch in pbar:
            x_batch, y_batch = x_batch.to(device), y_batch.to(device)
            y_target = y_batch.squeeze(0).cpu().numpy()

            x_mean = x_batch.mean(dim=(1, 2), keepdim=True)
            x_std = x_batch.std(dim=(1, 2), keepdim=True) + 1e-8
            x_batch_norm = (x_batch - x_mean) / x_std

            preds = model(x_batch_norm).cpu().numpy()

            # IC
            ic, _ = spearmanr(preds, y_target)
            if not np.isnan(ic):
                daily_ic.append(ic)

            # Top 10% 做多
            num_stocks = len(preds)
            k = max(1, int(num_stocks * 0.1))
            sorted_indices = np.argsort(preds)
            long_indices = sorted_indices[-k:]

            gross_return = np.mean(y_target[long_indices])
            market_return = np.mean(y_target)

            # 扣除手續費：每日報酬 = 毛報酬 - 交易成本
            net_return = gross_return - daily_cost

            if not np.isnan(gross_return):
                daily_gross_returns.append(gross_return)
                daily_net_returns.append(net_return)
                daily_benchmark_returns.append(market_return)

    # --- 計算統計指標 ---
    gross_arr = np.array(daily_gross_returns)
    net_arr = np.array(daily_net_returns)
    bench_arr = np.array(daily_benchmark_returns)

    # 年化報酬率
    ann_gross = np.mean(gross_arr) * 252
    ann_net = np.mean(net_arr) * 252
    ann_bench = np.mean(bench_arr) * 252
    ann_fee_drag = (np.mean(gross_arr) - np.mean(net_arr)) * 252  # 手續費年化拖累

    # 淨超額報酬
    net_excess = net_arr - bench_arr
    ann_net_excess = np.mean(net_excess) * 252

    # 夏普值（用淨報酬算才是真實的）
    sharpe_gross = (np.mean(gross_arr) / np.std(gross_arr)) * np.sqrt(252) if np.std(gross_arr) > 0 else 0
    sharpe_net = (np.mean(net_arr) / np.std(net_arr)) * np.sqrt(252) if np.std(net_arr) > 0 else 0

    # IR
    ir_net = (np.mean(net_excess) / np.std(net_excess)) * np.sqrt(252) if np.std(net_excess) > 0 else 0

    # 累積淨值曲線
    cumulative_net = np.cumprod(1 + net_arr)
    cumulative_bench = np.cumprod(1 + bench_arr)

    # 最大回撤 (Max Drawdown)
    peak = np.maximum.accumulate(cumulative_net)
    drawdown = (cumulative_net - peak) / peak
    max_drawdown = np.min(drawdown)

    # 勝率
    win_rate = np.mean(net_arr > 0) * 100

    # --- 輸出報告 ---
    print("\n" + "="*60)
    print("📈 純做多策略 Top 10% — 扣除手續費後績效報告")
    print("="*60)
    print(f"  平均 IC: {np.mean(daily_ic):.4f}")
    print(f"  交易天數: {len(net_arr)} 天")
    print("-"*60)
    print(f"  {'指標':<25} {'未扣手續費':>12} {'扣除手續費後':>12}")
    print("-"*60)
    print(f"  {'年化報酬率':<23} {ann_gross*100:>11.2f}% {ann_net*100:>11.2f}%")
    print(f"  {'夏普值 (Sharpe)':<23} {sharpe_gross:>12.2f} {sharpe_net:>12.2f}")
    print("-"*60)
    print(f"  市場基準年化報酬率: {ann_bench*100:.2f}%")
    print(f"  淨超額年化報酬 (Alpha): {ann_net_excess*100:.2f}%")
    print(f"  資訊比率 (IR): {ir_net:.2f}")
    print(f"  手續費年化拖累: {ann_fee_drag*100:.2f}%")
    print(f"  最大回撤 (MDD): {max_drawdown*100:.2f}%")
    print(f"  每日勝率: {win_rate:.1f}%")
    print(f"  期末累積淨值: {cumulative_net[-1]:.4f} (基準: {cumulative_bench[-1]:.4f})")
    print("="*60 + "\n")

    return np.mean(daily_ic), ann_net, sharpe_net


# --- 執行評估 ---
print("📊 測試集 (2023~2026) — 扣除手續費後績效：")
evaluate_with_fee(model, test_loader)
```
